```python
import cv2
from skimage.metrics import structural_similarity as ssim
```

```python
def load_images(image_paths):
    images = []
    for path in image_paths:
        img = cv2.imread(path)
        images.append(img)
    return images
```

```python
def calculate_ssim(img1, img2):
    gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
    gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
    score, diff = ssim(gray1, gray2, full=True)
    return score
```

```python
def remove_similar_images(images, threshold):
    unique_images = []
    for i in range(len(images)):
        is_unique = True
        for j in range(i + 1, len(images)):
            ssim_score = calculate_ssim(images[i], images[j])
            if ssim_score > threshold:
                is_unique = False
                break
        if is_unique:
            unique_images.append(images[i])
    return unique_images
```

```python
from PIL import Image
from pathlib import Path


def resize_images(input_dir, new_width, new_height):
    """Resizes all images in a directory to a specified size.
  
    Args:
      input_dir: Path to the input and output directory.
      new_width: Desired width of the resized images.
      new_height: Desired height of the resized images.
    """

    for image_file in Path(input_dir).glob('*.*'):
        img = Image.open(image_file)
        img = img.resize((new_width, new_height))
        img.save(image_file)


# Example usage:
input_dir = "data/extracted_frames"
new_width = 640
new_height = 640
try:
    resize_images(input_dir, new_width, new_height)
except:
    print()
```

```python
import os

image_dir = "data/extracted_frames"
image_paths = [os.path.join(image_dir, f) for f in os.listdir(image_dir) if f.endswith('.jpg')]

images = load_images(image_paths)
unique_images = remove_similar_images(images, 0.9)  # Adjust threshold as needed

# Save unique images or perform further processing
```

```python
import os


def remove_every_second_file(directory_path):
    """Removes every second file from the specified directory.
  
    Args:
      directory_path: The path to the directory.
    """

    files = os.listdir(directory_path)
    for i, file in enumerate(files):
        if i % 2 == 1:  # Every second file
            file_path = os.path.join(directory_path, file)
            try:
                os.remove(file_path)
                print(f"Removed: {file_path}")
            except OSError as e:
                print(f"Error removing file {file_path}: {e}")


if __name__ == "__main__":
    directory_path = "data/extracted_frames"
    remove_every_second_file(directory_path)
```
