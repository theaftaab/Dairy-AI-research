```python
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
DSET_DIR = "data/RF_sorted_data/"
```

```python
num_subdirs = len(os.listdir(DSET_DIR))
num_subdirs
```

```python
def get_subdir_files(directory) -> dict:
    data_files = {}
    for subdir in os.listdir(directory):
        subdir_files = []
        subdir_path = os.path.join(directory, subdir)
        for file in os.listdir(subdir_path):
            if file.endswith(".jpg"):
                subdir_files.append(file)
        data_files[subdir] = subdir_files
    return data_files
```

```python
dset_imgs = get_subdir_files(DSET_DIR)
dset_imgs['BCS-1.5']
```

```python
def plot_BCS_cnt(img_dset):
    data = {}
    for key in img_dset.keys():
        if key.startswith("BCS"):
            cnt = len(img_dset[key])
            data[key[4:]] = cnt

    sorted_data = dict(sorted(data.items()))
    x, y = sorted_data.keys(), sorted_data.values()
    plt.bar(x, y, color='teal')
    plt.grid(True, alpha=0.3)
    plt.xticks(rotation=45)
    plt.xlabel('BCS')
    plt.ylabel('No. of images')
```

```python
plot_BCS_cnt(dset_imgs)
```

```python
def plot_TScore_cnt(img_dset):
    data = {}
    for key in img_dset.keys():
        if key.startswith("Teat-Score"):
            cnt = len(img_dset[key])
            data[key[-1]] = cnt

    sorted_data = dict(sorted(data.items()))
    x, y = sorted_data.keys(), sorted_data.values()
    plt.bar(x, y, color='purple')
    plt.grid(True, alpha=0.3)
    plt.xticks()
    plt.xlabel('Teat Score')
    plt.ylabel('No. of images')
```
